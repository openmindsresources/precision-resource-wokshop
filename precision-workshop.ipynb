```python
# AS 1.0.0 - Data Preparation
import pandas as pd

# load datasets (users)
customers_df = pd.read_csv("./workshop/datasets/demo_customers.csv")
orders_df = pd.read_csv("./workshop/datasets/demo_orders.csv")
lineitems_df = pd.read_csv("./workshop/datasets/demo_lineitems.csv")

# set dates (users)
start_date = "2024-07-01"
end_date = "2024-12-31"
analysis_date = (pd.to_datetime(end_date) + pd.Timedelta(days=1)).strftime("%Y-%m-%d")

# display head data for each dataframe
print("Customers Data (First 5):")
print(customers_df.head(), "\n")
print("Orders Data (First 5):")
print(orders_df.head(), "\n")
print("Lineitems Data (First 5):")
print(lineitems_df.head(), "\n")
```
